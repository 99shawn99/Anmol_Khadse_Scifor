```python
#Pandas
#Pandas is mostly usd for data manipulation and analysis.
import pandas as pd

# Creating a DataFrame
data = {'Name': ['A', 'B', 'C'], 'Age': [25, 30, 35]}
df = pd.DataFrame(data)

# Displaying the DataFrame
print(df)

# Basic operations
print(df.describe())  # Summary statistics
print(df['Name'])     # Accessing a column
```

```python
#numpy
#mostly used for numerical operations and support array
import numpy as np

# Creating an array
arr = np.array([1, 2, 3, 4, 5])

# Basic operations
print(arr + 5)            # Element-wise addition
print(np.mean(arr))       # Mean of the array
print(np.random.rand(3))  # Random numbers
```

```python
#matplotlib
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Creating a line plot
plt.plot(x, y)
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
#seaborn
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = sns.load_dataset('iris')

# Creating a pair plot
sns.pairplot(data, hue='species')
plt.title("Pair Plot")
plt.show()
```
