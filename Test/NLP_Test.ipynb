```python
#importing necessary lib''s
import nltk
import spacy
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import wordnet
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from nltk import FreqDist
from collections import Counter
import string


# Download necessary NLTK data files
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')

# Load spaCy model
nlp = spacy.load("en_core_web_sm")
```

```python
#1-a)in sentences

story = """In the rugged terrain of the Western Ghats, Shivaji Maharaj, the founder of the Maratha Empire, was preparing for another day of strategic planning and military campaigns. Known for his innovative tactics and leadership, Shivaji Maharaj was determined to strengthen his kingdom and protect his people from external threats.

Shivaji Maharaj’s stronghold was the Raigad Fort, a formidable fortress perched high on a hilltop. The fort's strategic location allowed him to oversee the surrounding lands and defend against invaders. His trusted advisors and generals would gather in the fort's grand hall to discuss plans, share intelligence, and make critical decisions for the empire's expansion.

The evenings in Shivaji Maharaj's court were vibrant with activity. He would host feasts and gatherings to bolster the morale of his soldiers and celebrate victories. During these events, Shivaji Maharaj would address his people, sharing his vision for a prosperous and united Maratha Empire. His leadership inspired loyalty and courage among his subjects.

As night fell, Shivaji Maharaj would retire to his quarters, reflecting on the day's achievements and preparing for the challenges ahead. His dreams were filled with visions of a strong and resilient empire, where justice and valor would prevail."""

sentences = sent_tokenize(story)
print(sentences)
```

```python
#1-b)in words

words = word_tokenize(story)
print(words)
```

```python
#2)lowercasing

lowercased_story = story.lower()
print(lowercased_story)
```

```python
#3-a)Remove common stopwords.

def remove_stopwords(text):
    """Removes stopwords from a text string.

    Args:
      text: The text string to remove stopwords from.

    Returns:
      A new text string with stopwords removed.
    """

    stop_words = set(stopwords.words("english"))  # Import stopwords from nltk library
    word_tokens = word_tokenize(text)  # Import word_tokenize from nltk library

    filtered_text = [word for word in word_tokens if word not in string.punctuation]
    filtered_text = [word for word in filtered_text if word not in stop_words]

    return filtered_text

# Example usage
filtered_words = remove_stopwords(story)
filtered_story = ' '.join(filtered_words)
print(filtered_story)
```

```python
#3-b)List the removed stopwords.

removed_stopwords = [word for word in words if word.lower() in stopwords.words('english')]
print(removed_stopwords)
```

```python
#4)Punctuation Removal

def rem_punct(text):
    translator = str.maketrans('', '', string.punctuation)
    return text.translate(translator)

cleaned_story = rem_punct(story)
print(cleaned_story)
```

```python
#5-a) Apply stemming to the words in the story.

from nltk.stem.porter import PorterStemmer

stemmer = PorterStemmer()

def stem_words(text):
    word_tokens = word_tokenize(text)
    stems = [stemmer.stem(word) for word in word_tokens]
    return stems

stemmed_words = stem_words(filtered_story)
print(stemmed_words)
```

```python
#5-b)Provide a few examples of stemmed words.

example_stems = {word: stemmer.stem(word) for word in filtered_words[:10]}
print(example_stems)
```

```python
#6-a)Apply lemmatization to the words in the story.

def lemmatize_words(text):
    word_tokens = word_tokenize(text)
    lemmas = [wordnet.WordNetLemmatizer().lemmatize(word, pos='v') for word in word_tokens]
    return lemmas

lemmatized_words = lemmatize_words(filtered_story)
print(lemmatized_words)
```

```python
#6-b)Provide a few examples of lemmatized words.

example_lemmatized = {word: wordnet.WordNetLemmatizer().lemmatize(word, pos='v') for word in filtered_words[:10]}
print(example_lemmatized)
```

```python
#7)Named Entity Recognition (NER)

doc = nlp(story)
entities = [(ent.text, ent.label_) for ent in doc.ents]
print(entities)
```

```python
#8)Part-of-Speech Tagging

from nltk import pos_tag

def pos_tagg(text):
    word_tokens = word_tokenize(text)
    return pos_tag(word_tokens)

pos_tags = pos_tagg(story)
print(pos_tags[:10])
```

```python
#9)Term Frequency

term_freq_shivaji_maharaj = words.count("Shivaji")
print(term_freq_shivaji_maharaj)
```

```python
#10)Frequency Distribution
freqDist = FreqDist(words)
common_words = freqDist.most_common(10)
print(common_words)

# Plotting the frequency distribution
plt.figure(figsize=(10, 5))
freqDist.plot(10, title="Top 10 Most Common Words")
plt.show()
```

```python
#11)Token Count

token_count = len(words)
print(token_count)
```

```python
#12)Sentence Count

sentence_count = len(sentences)
print(sentence_count)
```

```python
#13)Stopword Frequency

stopword_freq = Counter(removed_stopwords)
top_stopwords = stopword_freq.most_common(3)
print(top_stopwords)
```

```python
#14)Longest Word

longest_word = max(words, key=len)
print(longest_word, len(longest_word))
```

```python
#15)Word Cloud

wordcloud = WordCloud().generate(story)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
#16)Noun Identification

nouns = [word for word, pos in pos_tags if pos.startswith('NN')]
print(nouns)
```

```python
#17)Verb Identification

verbs = [word for word, pos in pos_tags if pos.startswith('VB')]
print(verbs)
```

```python
#18)Word Context

empire_sentences = [sentence for sentence in sentences if 'empire' in sentence]
print(empire_sentences)
```

```python
#19)Named Entity Categorization

people = [ent for ent in entities if ent[1] == 'PERSON']
locations = [ent for ent in entities if ent[1] == 'GPE']
organizations = [ent for ent in entities if ent[1] == 'ORG']
print("People:", people)
print("Locations:", locations)
print("Organizations:", organizations)
```

```python
#20)Word Frequency Comparison

kingdom_count = words.count("kingdom")
feast_count = words.count("feast")
print("Kingdom:", kingdom_count)
print("Feast:", feast_count)
```
